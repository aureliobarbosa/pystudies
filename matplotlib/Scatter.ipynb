```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
```

Let's create some random numbers.

```python
points = 100
rng = np.random.RandomState(0)
x =  rng.randn(points)
y =  rng.randn(points)
colors = rng.rand(points)
sizes =  1000*rng.rand(points)
```

```python
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar();
plt.axis([-3,3,-3,3]);
```

Only to check whether I understood that, let's use another color map.

```python
cmap = mpl.colors.LinearSegmentedColormap.from_list('my_colormap', ['yellow','red'], 256)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap=cmap)
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar();
plt.axis([-3,3,-3,3]);
```

To finish, and check that I understood Object Oriented Plotting, let's it this way.

```python
cmap = mpl.colors.LinearSegmentedColormap.from_list('my_colormap', ['yellow','red'], 256)
```

```python
plt.style.use('classic')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2,sharey=True,figsize=(9,4))

#Making the first figure
img = ax[0].scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap=cmap)
ax[0].set(xlim=[-3,3],xlabel='x',ylabel='y',title='r1')

ax[1].scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap=cmap)
ax[1].set(xlim=[-3,3],xlabel='x',title='r2');

cax = plt.axes([0.9,0.1,0.025,0.75])
#plt.subplots_adjust(left=0.1,right=0.8,bottom=0.1,top=0.9)
plt.subplots_adjust(right=0.85,bottom=0.1, top=0.85)
#plt.colorbar(img,cmap=cmap,ax=ax[1]);
plt.colorbar(img,cmap=cmap,cax=cax);

```
