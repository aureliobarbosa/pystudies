Simple plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
plt.style.use('seaborn-whitegrid')
```

Object oriented plotting

```python
x = np.linspace(0,10,1000)
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x,np.cos(x))
```

Everything can be done on the background, to create simple plots.

```python
plt.plot(x,np.cos(x))
```

Now, for plotting two curves in a single plot, using OO Plotting

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.cos(x))
ax.plot(x, np.sin(x))
```

A little bit of more control in our graphic.

```python
fig = plt.figure()
ax = plt.axes()

ax.set_title('Trigonometric Functions')
ax.plot(x, np.cos(x),'-g',label='cos(x)')
ax.plot(x, np.sin(x),':r',label='sin(x)')
ax.legend()
```

The same figure can be made using the property set. Note that there is a distinction in MATLAB-Style and object oriented style. This is explained on page 232 of Python Data Science Handbook. 

```python
ax = plt.axes()
x = np.linspace(-5,5,100)
ax.plot(x, np.cos(x),'-g',label='cos(x)')
ax.plot(x, np.sin(x),':r',label='sin(x)')
ax.set( title = 'Trigonometric Fuctions', xlabel = 'x', ylabel = 'y',
       xlim = (-5,5), ylim=(-2,2));
```
