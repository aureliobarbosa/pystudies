```python
from sympy import lambdify, sympify, diff
import numpy as np
import matplotlib.pyplot as plt
```

## Two variable function.

```python
test      = sympify('exp(-(x+y))/(1.1+sin(x**2+y**2))')
dxtest    = diff(test,'x')
dytest    = diff(test,'y')
dxdytest  = diff(dytest,'x')

np_test   = lambdify(('x','y'), test)
np_dxtest = lambdify(('x','y'), dxtest)
np_dytest = lambdify(('x','y'), dytest)
np_dxdy   = lambdify(('x','y'), dxdytest)
```

```python
x = np.linspace(0.0,2.0*np.pi,num=100)
y = np.linspace(0.0,2.0*np.pi,num=100)

xx,yy = np.meshgrid(x,y)

z1 = np_test(xx,yy)

z2 = np_dxtest(xx,yy)
z3 = np_dytest(xx,yy)
z4 = np_dxdy(xx,yy)
```
