# Basic Plotting - circle

**This notebook** shows how to use numpy and matplotlib to draw a circle and save it to a pdf file that can be used latter in a pdflatex environment!

---


```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
```

Making a **CIRCLE** with diameter `r`

```python
y = np.linspace(0,1,101)
print(y)
```

```python
r = 2.0
theta = np.linspace(0.0,2*pi,100)
x = r*np.cos(theta)
y = r*np.sin(theta)
```

```python
dr = 0.1

plt.figure()
plt.axis([-r-dr,r+dr,-r-dr,r+dr])
plt.legend(['points'])                     # Note that legend do not appear because there is only one group of points
plt.xlabel('x')
plt.ylabel('y')
plt.title('Circle')
plt.plot(x,y)
plt.savefig('circle.pdf')
```
