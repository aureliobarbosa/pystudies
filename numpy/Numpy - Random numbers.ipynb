# Random Numbers in numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

Seed

```python
np.random.seed(125)
```

Random numbers - flat distribution

Single point and (x,y)

```python
print(np.random.rand())
print(np.random.rand(2,10))
```

# Gráfico 2D distribuição uniforme

```python
# x = np.random.rand(1000)
# y = np.random.rand(1000)

x, y = np.random.rand(2, 1000) # more pythonic

plt.figure(1)
plt.title("Números aleatórios")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y, 'bo')
```

# Gráfico 2D -  Distribuição normal

```python
x , y = np.random.randn(2, 1000)

plt.figure(2)
plt.title("Números aleatórios com dist. normal ")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y, 'bo')
```

# Grafico - 2D distribuição de inteiros

```python
x , y = np.random.randint(101,  size = (2, 100))

plt.figure(2)
plt.title("Números inteiros aleatórios com dist. normal ")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y, 'bo')
```
