# Basic Plotting - Ellipse

This notebook shows how to use numpy and matplotlib to draw an ellipse and save it to a pdf file that can be used latter in a pdflatex environment!

```python
%reset
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi as pi
```

Making a **ELLIPSE** with major-axis `a=1.0` along the `x` direction.

The eccentricity is defined by the variable `ecc`.

```python
a = 1.0
ecc = 0.4
b = np.sqrt(1-ecc**2)
print(b)
```

```python
theta = np.linspace(0.0,2*pi,100)
x = a*np.cos(theta)
y = b*np.sin(theta)
```

```python
da = a/10
db = b/10
plt.figure()
plt.axis([-a-da,a+da,-b-db,b+db])
plt.legend(['points'])                     # Note that legend do not appear because there is only one group of points
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ellipse')
plt.plot(x,y)
plt.savefig('ellipse.pdf')
```
